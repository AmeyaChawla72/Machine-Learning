```python
def h(theta,x):
    b=theta[0]+theta[1]*x[0]+theta[2]*x[1]+theta[3]*x[2]
    return sigmoid(b)
```

```python
def sigmoid(z):
    return 1.0/(1.0+np.exp(-1*z))
```

```python
import math
def error(theta,x,y):
    m=x.shape[0]
    e=0.0
    for i in range(m):
        e+=(  y[i]*math.log(h(theta,x[i])) + (1-y[i])*math.log(1-h(theta,x[i]))  )
        
    return -1*e/m
        
        
```

```python
def gradient(theta,x,y):
    grad=np.zeros((4,))
    m=x.shape[0]
    for i in range(m):
        g=h(theta,x[i])
        grad[0]+=(g-y[i])
        grad[1]+=(g-y[i])*x[i][0]
        grad[2]+=(g-y[i])*x[i][1]
        grad[3]+=(g-y[i])*x[i][2]
    return grad/m
```

```python
def theta(x,y):
    theta=np.zeros((4,))
    m=600
    e=[]
    for i in range(m):
        g=gradient(theta,x,y)
        for j in range(4):
            theta[j]-=0.02*g[j]
        e.append(error(theta,x,y))
            
    return theta,e
```
